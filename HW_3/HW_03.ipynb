```python
from sklearn.datasets import load_iris

iris = load_iris()

from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier 

k_range = range(1, 31)

cv_scores = []

for k in k_range:
    
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X = iris.data, y = iris.target, cv = 10)
    cv_scores.append(scores.mean())

import matplotlib.pyplot as plt


plt.plot(k_range, cv_scores, color='blue')
plt.title('KNN Hyperparameter')
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()

k_range = range(11, 16) 
cv_scores = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X = iris.data, y = iris.target, cv = 10)
    cv_scores.append(scores.mean())

plt.plot(k_range, cv_scores, color='blue')
plt.title('KNN Hyperparameter K 11-15')
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()

from sklearn.model_selection import GridSearchCV

X = iris.data
Y = iris.target



param = {'n_neighbors' : [13]}

grid = GridSearchCV(estimator=KNeighborsClassifier(), param_grid=param, cv=10, scoring='accuracy')

grid.fit(X, Y)

print(f"Best k value (from GridSearchCV): {grid.best_params_['n_neighbors']}")
print(f"Best score (from GridSearchCV): {grid.best_score_}")




```
